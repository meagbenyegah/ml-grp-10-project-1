```python
# ============================================================
# CSCD613 - Machine Learning with Big Data Analytics (Group 10)
# Project 1
# 04_Model_Interpretation_and_Conclusion.ipynb
# ------------------------------------------------------------
# Description:
# This notebook interprets the trained XGBoost model, explains
# key findings from feature importance, evaluates business implications,
# and summarizes the overall conclusions from the analysis.
# ============================================================

# -------------------------
# 1. Import Dependencies
# -------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import shap

from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')

# -------------------------
# 2. Load Data and Model
# -------------------------
data = pd.read_csv("data/featured_application_train.csv")
X = data.drop("TARGET", axis=1)
y = data["TARGET"]

best_model = joblib.load("models/xgboost_model.pkl")

print("Model and data loaded successfully.")

# -------------------------
# 3. Feature Importance Review
# -------------------------
importances = best_model.feature_importances_
feature_names = X.columns

feat_imp = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feat_imp = feat_imp.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feat_imp.head(15), palette='mako')
plt.title("Top 15 Most Influential Features")
plt.tight_layout()
plt.show()

print("\nTop 10 Features Influencing Loan Default Prediction:")
print(feat_imp.head(10))

# -------------------------
# 4. SHAP Value Interpretation
# -------------------------
# SHAP helps explain how each feature contributes to a models prediction
explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X)

# Global explanation summary
shap.summary_plot(shap_values, X, plot_type="bar", show=False)
plt.title("SHAP Feature Importance Summary (Global Impact)")
plt.show()

# Detailed impact plot
shap.summary_plot(shap_values, X, show=False)
plt.title("SHAP Summary Plot (Feature Influence Direction)")
plt.show()

# -------------------------
# 5. Individual Prediction Explanation
# -------------------------
# Pick a random instance for local interpretability
sample_index = np.random.randint(0, len(X))
sample = X.iloc[sample_index]

print(f"\nInterpreting prediction for sample index: {sample_index}")
shap.force_plot(
    explainer.expected_value, 
    shap_values[sample_index, :], 
    sample,
    matplotlib=True
)

# -------------------------
# 6. Insights & Interpretation
# -------------------------
print("\n=== INTERPRETATION ===")
print("""
The model identifies that:
- Income-related variables (AMT_INCOME_TOTAL, ANNUITY_INCOME_RATIO) are strong predictors.
- Credit size and duration indicators (AMT_CREDIT, AMT_ANNUITY) significantly affect loan risk.
- Demographic features like education level and employment years also contribute meaningfully.
- Features related to phone change history and region population reflect socio-economic stability.

High SHAP values indicate that:
- Applicants with higher credit-to-income ratios tend to be at higher default risk.
- Pensioners and low-income groups show slightly higher risk probability.
""")

# -------------------------
# 7. Business Implications
# -------------------------
print("\n=== BUSINESS IMPLICATIONS ===")
print("""
1. The model can assist banks and microfinance institutions in automating loan approval risk scoring.
2. Clients with high loan-to-income ratios or inconsistent employment histories should be flagged for manual review.
3. The model improves fairness by relying on quantitative data rather than manual judgment.
4. Financial institutions can reduce default rates by integrating this predictive system into loan management workflows.
""")

# -------------------------
# 8. Limitations and Future Work
# -------------------------
print("\n=== LIMITATIONS AND FUTURE WORK ===")
print("""
- The dataset may not fully represent all financial institutions in Ghana; model generalizability needs further testing.
- Missing data handling and encoding could introduce bias.
- Future work should include additional behavioral and transactional data for deeper insights.
- Testing on real-time application data will help validate robustness.
""")

# -------------------------
# 9. Conclusion
# -------------------------
print("\n=== CONCLUSION ===")
print("""
This project successfully applied machine learning (XGBoost)
to predict loan default probability using applicant and credit-related features.

The model achieved high predictive performance with interpretable results.
Insights derived from SHAP and feature importance can help financial institutions
make more data-driven and equitable lending decisions in Ghanas financial sector.

 End of Project 1  Machine Learning with Big Data Analytics (Group 10)
""")
```
